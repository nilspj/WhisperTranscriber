Komórka poniżej zawiera dwie linie zaczynające się od „input_filename=" i „output_filename=". Wpisz nazwę twojego pliku miedzy dwie cudzysłowie po znaku równości w pierwszei linji.
```python
input_filename = "Marcin 41.mp4"
output_filename = "Marcin 41.txt"
```

#5 Rozpocznij transkrypcję
Kliknij przycisk odtwarzania w lewym górnym rogu komórki poniżej. Wokół przycisku odtwarzania pojawi się okrąg. Poczekaj, aż przestanie się obracać (10-30min). Nazwa ustalona w poprzednim kroku pojawi się w oknie Pliki po lewej stronie. To jest Twoja transkrypcja. Najedź na nią myszką i kliknij trzy kropki, które pojawią się po prawej stronie, a następnie w nowym oknie kliknij Pobierz, aby pobrać transkrypcję.

```python
import whisper
import numpy as np
import torch
torch.cuda.empty_cache()
model = whisper.load_model("large")
result = model.transcribe(input_filename)
#print(result["text"])
with open(output_filename, "w", encoding="utf-8") as txt:
    txt.write(result["text"])
torch.cuda.empty_cache()
```
